# PubMed con errores

```python
import pandas as pd

pubmed=pd.read_csv('pubmed_raw_corpus.csv')
pubmed
```

# 1) Formato de la fecha

```python
pubmed['Date'] = pd.to_datetime(pubmed['Date'], format='%Y %b %d', errors='coerce')
pubmed['Date'] = pubmed['Date'].dt.strftime('%d/%m/%Y')
pubmed
```

Guardar el archivo

```python
pubmed.to_csv('pubmed_clean_corpus.csv', index=False)
print("El archivo 'pubmed_clean_corpus.csv' ha sido guardado.")
```

# Arxiv con errores

```python
arxiv=pd.read_csv('/content/arxiv_papers.csv')
arxiv
```

# 1) Formato de la fecha

```python
arxiv['Date'] = arxiv['Date'].str.extract(r'Submitted on (\d{1,2} \w{3} \d{4})')

arxiv['Date'] = pd.to_datetime(arxiv['Date'], format='%d %b %Y', errors='coerce')
arxiv['Date'] = arxiv['Date'].dt.strftime('%d/%m/%Y')
arxiv
```

# 2) Eliminación del texto redundante

```python
arxiv['Title'] = arxiv['Title'].str.replace(r'^Title:\s*', '', regex=True)
arxiv['Authors'] = arxiv['Authors'].str.replace(r'^Authors:\s*', '', regex=True)
arxiv['Abstract'] = arxiv['Abstract'].str.replace(r'^Abstract:\s*', '', regex=True)

arxiv
```

Guardar el archivo

```python
arxiv.to_csv('arxiv_clean_corpus.csv', index=False)
print("El archivo 'arxiv_clean_corpus.csv' ha sido guardado.")
```
