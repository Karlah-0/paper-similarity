# PubMed con errores

```python
import pandas as pd

pubmed=pd.read_csv('../data/pubmed_raw_corpus.csv')
pubmed.head()
```

# 1) Formato de la fecha

```python
pubmed['Abstract'] = pubmed['Abstract'].str.replace('\n',' ')
pubmed['Abstract'] = pubmed['Abstract'].str.replace('\r',' ')
pubmed['Abstract'] = pubmed['Abstract'].str.replace('\t',' ')
pubmed['Abstract'] = pubmed['Abstract'].str.replace('     ',' ')
pubmed['Title'] = pubmed['Title'].str.replace('\n',' ')
pubmed['Title'] = pubmed['Title'].str.replace('\r',' ')
pubmed['Title'] = pubmed['Title'].str.replace('\t',' ')
pubmed['Title'] = pubmed['Title'].str.replace('     ',' ')
pubmed.head()
```

```python
def fill_missing_date(date):
    try:
        return pd.to_datetime(date, format='%Y %b %d').strftime('%d/%m/%Y')
    except:
        try:
            return pd.to_datetime(date, format='%Y %b').strftime('00/%m/%Y')
        except:
            try:
                return pd.to_datetime(date, format='%Y').strftime('00/00/%Y')
            except:
                return '00/00/0000'
pubmed['Date'] = pubmed['Date'].apply(fill_missing_date)
pubmed.head()
```

Guardar el archivo

```python
pubmed.to_csv('../data/pubmed_clean_corpus.csv', index=False)
print("El archivo 'pubmed_clean_corpus.csv' ha sido guardado.")
```

# ArXiv con errores

```python
arxiv = pd.read_csv('../data/arxiv_raw_corpus.csv')
arxiv.head()
```

```python
for i in range(5):
    print(arxiv.at[i, 'Date'])
```

```python
arxiv['Date'] = arxiv['Date'].str.extract(r'(\d{1,2} \w{3} \d{4})')
arxiv['Date'] = pd.to_datetime(arxiv['Date'], format='%d %b %Y', errors='coerce')
arxiv['Date'] = arxiv['Date'].dt.strftime('%d/%m/%Y')
arxiv.head()
```

```python
arxiv.to_csv('../data/arxiv_clean_corpus.csv', index=False)
print("El archivo 'arxiv_clean_corpus.csv' ha sido guardado.")
```
